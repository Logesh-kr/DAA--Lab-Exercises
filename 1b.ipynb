```python
# Interpolation Search

def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        if low == high:
            if arr[low] == key:
                return low, probes
            return -1, probes

        pos = low + ((key - arr[low]) * (high - low)) // (arr[high] - arr[low])

        if arr[pos] == key:
            return pos, probes

        if arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


# Binary Search

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1

        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, probes

        if arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


# Sample Roll Numbers
roll_numbers = list(range(1, 10001))

key = int(input("Enter roll number to search: "))

idx1, probes1 = interpolation_search(roll_numbers, key)
idx2, probes2 = binary_search(roll_numbers, key)

print("\nInterpolation Search")
print("Position:", idx1)
print("Probes:", probes1)

print("\nBinary Search")
print("Position:", idx2)
print("Probes:", probes2)
```
