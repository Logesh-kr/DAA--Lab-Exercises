```python
import random


def interpolation_search_float(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= key <= arr[high]:

        comparisons += 1

        if arr[high] == arr[low]:
            break

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        if abs(arr[pos] - key) < 1e-6:
            return comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return comparisons


sizes = [10000, 50000, 100000]

print("Dataset Size\tComparisons")

for n in sizes:

    data = sorted(random.uniform(0.0, 1000.0) for _ in range(n))

    target = data[n // 2]

    comparisons = interpolation_search_float(data, target)

    print(f"{n}\t\t{comparisons}")
```
